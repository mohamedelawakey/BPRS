# New Features

```python
# drop the id column cause it's not useful
data = data.drop(['Id'], axis=1)
```

```python
data.head()
```

```python
"""
edit columns 'RatingDist5', 'RatingDist4', 'RatingDist3', 'RatingDist2', 'RatingDist1' 
to extract only the number from them 5:3 to 3 in all these features

edit the column of RatingDistTotal to covert it from total:3 to 3 only 

create a new feature that calc the average of rating of the 5 columns
"""

# function to extract numbers only from the 5 columns 
def extract_numbers_only_from_ratings_1_to_5_columns(value: str) -> int:
    if pd.isna(value) or not isinstance(value, str):
        return 0

    if ':' in value:
        parts = value.split(':')
        if len(parts) == 2:
            try:
                return int(parts[1].strip())
            except:
                return 0
    return 0

# function to extract the number only from total ratings
def extract_numbers_only_from_total_ratings(value: str) -> int:
    if pd.isna(value) or not isinstance(value, str):
        return 0

    return int(value.replace('total:', '').strip()) if 'total:' in value else 0
```

```python
# apply the function to columns

rating_col_1_to_5 = ['RatingDist5', 'RatingDist4', 'RatingDist3', 'RatingDist2', 'RatingDist1']

for col in rating_col_1_to_5:
    data[col] = data[col].apply(extract_numbers_only_from_ratings_1_to_5_columns)
    
data['RatingDistTotal'] = data['RatingDistTotal'].apply(extract_numbers_only_from_total_ratings)
```

```python
data.head()
```

```python
# new feature which calc the weighted average rating

all_sum  = (data['RatingDist5'] * 5) + (data['RatingDist4'] * 4) + (data['RatingDist3'] * 3) + (data['RatingDist2'] * 2) + (data['RatingDist1'] * 1)
data['weighted_rating'] = all_sum / data['RatingDistTotal'].replace(0, 1).round(2)
```

```python
# new feature that calc the percentage of ratings
"""
calculate the average of each column in the ratings columns
calculate the high average of high ratings columns
calculate the low average of low ratings columns
"""

data['average_rating_5'] = (data['RatingDist5'] / data['RatingDistTotal'].replace(0, 1)) * 100
data['average_rating_4'] = (data['RatingDist4'] / data['RatingDistTotal'].replace(0, 1)) * 100
data['average_rating_2'] = (data['RatingDist2'] / data['RatingDistTotal'].replace(0, 1)) * 100
data['average_rating_1'] = (data['RatingDist1'] / data['RatingDistTotal'].replace(0, 1)) * 100
data['average_high_rating'] = ((data['RatingDist5'] + data['RatingDist4']) / data['RatingDistTotal'].replace(0, 1)) * 100
data['average_low_rating'] = ((data['RatingDist1'] + data['RatingDist2']) / data['RatingDistTotal'].replace(0, 1)) * 100

data['average_rating_5'] = data['average_rating_5'].round(2)
data['average_rating_4'] = data['average_rating_4'].round(2)
data['average_rating_2'] = data['average_rating_2'].round(2)
data['average_rating_1'] = data['average_rating_1'].round(2)
data['average_high_rating'] = data['average_high_rating'].round(2)
data['average_low_rating'] = data['average_low_rating'].round(2)
```

```python
# convert the types of the columns to save the memory

int_columns = data.select_dtypes(include='int64').columns
float_columns = data.select_dtypes(include='float64').columns

for col in int_columns:
    data[col] = data[col].astype('int16')

for col in float_columns:
    data[col] = data[col].astype('float16')
    
data['PublishYear'] = data['PublishYear'].astype('int16')
```
